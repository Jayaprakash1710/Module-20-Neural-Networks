```python
df=pd.read_csv('spiral.csv')
```

```python
df.head()
```

```python
plt.scatter(df['x1'],df['x2'],c=df['y'])
```

```python
x=df.loc[::,'x1':'x2'].to_numpy()
```

```python
y=df['y'].to_numpy()
```

Shapes:

```python
m=x.shape[0]
n=x.shape[1]
p=len(np.unique(y))
```

```python
print(m,n,p)
```

1. Initializing w and b

```python
w=0.01*np.random.randn(n,p)
b=np.zeros((1,p))
```

```python
print(w.shape)
print(b.shape)
```

2. Forward pass

```python
z=np.dot(x,w)+b
```

```python
z.shape
```

3. Softmax:

```python
exp_z=np.exp(z)
```

```python
probs=exp_z/np.sum(exp_z,axis=1,keepdims=True)
```

4. Loss calculation

```python
error=-np.log(probs[range(m),y])
```

```python
total_error=np.sum(error)
mean_error=total_error/len(error)
print(mean_error)
```

5. Prediction

```python
y_hat=np.argmax(probs,axis=1)
print(y_hat)
```

```python
plt.scatter(x=df['x1'],y=df['x2'],c=y_hat)
```

6. Evaluation

```python
correct_predictions=np.sum(y==y_hat)
```

```python
accuracy=correct_predictions/len(y_hat)*100
```

```python
print(accuracy)
```
